**Lectura de los datos**

```python
dp=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
dp
```

**1. Filtro por objeto**

```python
objeto=dp[(dp["NOMBRE_CLIENTE"] == "VINILOS")]
objeto
```

**2. Filtro por filas**

```python
filas=dp.iloc[2:6,:]
filas
```

**3. Filtro por columnas**

```python
columnas=dp.iloc[:,0:4]
columnas
```

**4. Filtro de filas por columnas**

```python
dp1=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx",index_col="CANT")
dp1=dp1.loc[[1,2],["NOMBRE_CLIENTE", "PRECIO_UNITARIO", "COSTO_UNITARIO"]]
dp1
```

**5. Filtro por cabecera**

```python
cabecera=dp.head(10)
cabecera
```

**6. Filtro por cola**

```python
cola=dp.tail(10)
cola
```

**7. Filtro por comparación**

```python
comparacion=dp[dp["PRECIO_UNITARIO"]>20000]
comparacion
```

**8. Filtro Y (Conjunción)**

```python
conjuncion=dp[(dp["NOMBRE_VENDEDOR"]=="LETICIA RAMIREZ HERNANDEZ") & (dp["NOMBRE_CLIENTE"]=="PÚBLICO EN GENERAL") & (dp["SUBTOTAL_PARTIDA"]>4000)]
conjuncion
```

**9. Filtro O (Disyunción)**

```python
disyuncion=dp[(dp["CANT"]>15) | (dp["PRECIO_UNITARIO"]>10000)]
disyuncion
```

**10. Filtro Not (Negación)**

```python
negacion=dp[~(dp["COSTO_UNITARIO"]==0)]
negacion
```
